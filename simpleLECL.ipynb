```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY") 
 

```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.1-8b-instant",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Translate the following text from English to Telugu"),
    HumanMessage(content="Hello, how are you?")
    ]
result=model.invoke(messages)
```

```python
#only to output the tranlated message from AI Message
from langchain_core.output_parsers import StrOutputParser   
parser=StrOutputParser()
parser.invoke(result)
```

```python
## using LCEL-CHain the components
chain=model|parser
chain.invoke(messages)
```

```python
##prompt template
from langchain_core.prompts import ChatPromptTemplate
generic_template="Translate the following into {language}:"
prompt=ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
)
```

```python
result=prompt.invoke({"language":"Telugu","text":"did you eat anything"})
result.to_messages()
```

```python
chain=prompt|model|parser
chain.invoke({"language":"Telugu","text":"did you eat anything"})
```
